Caribbean Sea populaition

To create a radar graph (also known as a spider chart) to show the population density of countries bordering the Caribbean Sea, and then create a line graph to show the range and number of populations

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data for countries bordering the Caribbean Sea
countries = ['Cuba', 'Jamaica', 'Dominican Republic', 'Puerto Rico', 'Haiti', 'Bahamas']
population_density = [106, 265, 223, 400, 400, 39]  # Hypothetical population density (people per km)

# Radar Graph
# Number of variables
num_vars = len(countries)

# Compute angle for each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# The plot is a circle, so we need to "complete the loop" and append the start to the end.
population_density += population_density[:1]
angles += angles[:1]

# Create the radar chart
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)
ax.fill(angles, population_density, color='cyan', alpha=0.25)
ax.plot(angles, population_density, color='blue', linewidth=2)

# Labels for each axis
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(countries)

plt.title('Population Density of Caribbean Sea Countries (people per km)')
plt.show()

# Line Graph
# Hypothetical population data in millions for the same countries
population = [11.3, 2.9, 10.5, 3.2, 11.1, 0.4]  # Total population (hypothetical)

# Create the line graph
plt.figure(figsize=(10, 6))
plt.plot(countries, population, marker='o', color='orange', label='Population (millions)')

# Adding titles and labels
plt.title('Population of Caribbean Sea Countries')
plt.xlabel('Countries')
plt.ylabel('Population (millions)')
plt.xticks(rotation=45)  # Rotate country names for better visibility
plt.legend()
plt.grid()

# Show the plot
plt.show()
```
